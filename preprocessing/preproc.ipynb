```python
from pathlib import Path
import pandas as pd

# Project root (one level above this notebook's folder)
ROOT = Path.cwd().resolve().parent

player_csv = ROOT / 'data' / 'processed' / '2025-2026_player_stats_by_gameweek_all_clean.csv'
team_csv = ROOT / 'data' / 'processed' / '2025-2026_team_stats_by_gameweek_long_clean.csv'

df_player_2526 = pd.read_csv(player_csv)
df_team_2526 = pd.read_csv(team_csv)

print(f'Player data shape: {df_player_2526.shape}')
print(f'Team data shape: {df_team_2526.shape}')

# Quick preview for exploration
display(df_player_2526.head())
display(df_team_2526.head())
```

```python
import pandas as pd
from pathlib import Path

  # Works whether notebook is run from repo root or preprocessing/
CWD = Path.cwd().resolve()
BASE = CWD / "data" / "processed"
if not BASE.exists():
    BASE = CWD.parent / "data" / "processed"

print("Using base path:", BASE)

teams = pd.read_csv(BASE / "teams.csv")
code_to_name = teams.set_index("code")["name"].to_dict()

team_stats_path = BASE / "2025-2026_team_stats_by_gameweek_long_clean.csv"
df_team = pd.read_csv(team_stats_path)

df_team["team_name"] = pd.to_numeric(df_team["team"], errors="coerce").map(code_to_name)
df_team.to_csv(team_stats_path, index=False)

print("Saved:", team_stats_path)
print("Unmapped rows:", int(df_team["team_name"].isna().sum()))
df_team[["team", "team_name", "gameweek", "season"]].head()
```

```python
# === PLAYER DROP CELL (TEAM-MAP SAFE) ===
# Keeps keys needed to map to team stats: name, gameweek, season
player_merge_keys = ['name', 'gameweek', 'season']

drop_cols_player = [
    'id', 'first_name', 'second_name', 'status', 'news', 'fotmob_name', 'pulse_id',
    'now_cost', 'selected_by_percent', 'selected_rank', 'selected_rank_type',
    'cost_change_event', 'cost_change_event_fall', 'cost_change_start', 'cost_change_start_fall',
    'transfers_in_event', 'transfers_out_event', 'transfers_in', 'transfers_out',
    'now_cost_rank', 'now_cost_rank_type', 'form_rank', 'form_rank_type',
    'points_per_game_rank', 'points_per_game_rank_type',
    'influence_rank', 'influence_rank_type', 'creativity_rank', 'creativity_rank_type',
    'threat_rank', 'threat_rank_type', 'ict_index_rank', 'ict_index_rank_type',
    'corners_and_indirect_freekicks_order', 'direct_freekicks_order', 'penalties_order',
    'gw', 'source', 'team_code',
    'form', 'value_form', 'value_season', 'ep_next', 'ep_this', 'points_per_game',
    'chance_of_playing_next_round', 'chance_of_playing_this_round',
    'web_name',
    'strength', 'strength_overall_home', 'strength_overall_away',
    'strength_attack_home', 'strength_attack_away',
    'strength_defence_home', 'strength_defence_away'
]

drop_cols_player = [c for c in drop_cols_player if c not in player_merge_keys]
drop_cols_existing = [c for c in drop_cols_player if c in df_player_2526.columns]
df_player_2526_model = df_player_2526.drop(columns=drop_cols_existing).copy()

print('Dropped columns:', len(drop_cols_existing))
print('Original shape:', df_player_2526.shape)
print('Model shape:', df_player_2526_model.shape)
print('Merge keys present:', {k: (k in df_player_2526_model.columns) for k in player_merge_keys})
df_player_2526_model[['name', 'gameweek', 'season']].head()
```

```python
# === MERGE PLAYER + TEAM STATS CELL ===
# expects: df_player_2526_model (cleaned player df)
# expects: df_team (team stats df with team_name already mapped from team code)

df_p = df_player_2526_model.copy()
df_t = df_team.copy()

# Normalize join keys
df_p['name'] = df_p['name'].astype(str).str.strip()
df_t['team_name'] = df_t['team_name'].astype(str).str.strip()

df_p['gameweek'] = pd.to_numeric(df_p['gameweek'], errors='coerce')
df_t['gameweek'] = pd.to_numeric(df_t['gameweek'], errors='coerce')

df_p['season'] = df_p['season'].astype(str).str.strip()
df_t['season'] = df_t['season'].astype(str).str.strip()

# Aggregate team rows per team-week-season (handles duplicates, e.g. DGW sources)
team_keys = ['team_name', 'gameweek', 'season']
num_cols = df_t.select_dtypes(include='number').columns.tolist()
num_cols = [c for c in num_cols if c != 'gameweek']
df_t_agg = df_t.groupby(team_keys, as_index=False)[num_cols].mean()

# Merge: player's team name + gameweek + season
df_full_2526 = df_p.merge(
    df_t_agg,
    left_on=['name', 'gameweek', 'season'],
    right_on=['team_name', 'gameweek', 'season'],
    how='left',
    suffixes=('', '_team')
)

print('Player shape:', df_p.shape)
print('Team agg shape:', df_t_agg.shape)
print('Merged shape:', df_full_2526.shape)
print('Rows without team match:', int(df_full_2526['team_name'].isna().sum()))
df_full_2526.head()
```

```python
# === DROP PER-90 COLUMNS CELL ===
# Remove per-90 features (not needed for single-game prediction rows)

per90_cols = [c for c in df_full_2526.columns if '_per_90' in c]
df_full_2526_no_per90 = df_full_2526.drop(columns=per90_cols).copy()

print('Per-90 columns dropped:', len(per90_cols))
print(per90_cols)
print('Before:', df_full_2526.shape)
print('After :', df_full_2526_no_per90.shape)
df_full_2526_no_per90.head()
```

```python
# === SAVE FINAL DATAFRAME CSV CELL ===
# Save final preprocessed dataframe to CSV
from pathlib import Path

CWD = Path.cwd().resolve()
BASE = CWD / 'data' / 'processed'
if not BASE.exists():
    BASE = CWD.parent / 'data' / 'processed'

out_path = BASE / '2025-2026_player_team_merged_no_per90.csv'
df_full_2526_no_per90.to_csv(out_path, index=False)

print('Saved:', out_path)
print('Shape:', df_full_2526_no_per90.shape)
```

```python
# === SELF-CONTAINED: ADD SIDE USING CSVs ===
import pandas as pd
from pathlib import Path

# Resolve data/processed path robustly
CWD = Path.cwd().resolve()
BASE = CWD / 'data' / 'processed'
if not BASE.exists():
    BASE = CWD.parent / 'data' / 'processed'

merged_path = BASE / '2025-2026_player_team_merged_no_per90.csv'
team_stats_path = BASE / '2025-2026_team_stats_by_gameweek_long_clean.csv'

# Load directly from CSVs
df_merged = pd.read_csv(merged_path)
df_team = pd.read_csv(team_stats_path)

# Ensure team_name exists in team stats
if 'team_name' not in df_team.columns:
    teams = pd.read_csv(BASE / 'teams.csv')
    code_to_name = teams.set_index('code')['name'].to_dict()
    df_team['team_name'] = pd.to_numeric(df_team['team'], errors='coerce').map(code_to_name)

# Normalize keys
for d, team_col in [(df_merged, 'name'), (df_team, 'team_name')]:
    d[team_col] = d[team_col].astype(str).str.strip()
    d['gameweek'] = pd.to_numeric(d['gameweek'], errors='coerce')
    d['season'] = d['season'].astype(str).str.strip()

# Build side map only for unambiguous keys
keys = ['team_name', 'gameweek', 'season']
tmp = df_team[keys + ['side']].dropna(subset=['team_name', 'gameweek', 'season']).copy()

side_counts = (
    tmp.groupby(keys, as_index=False)['side']
    .nunique()
    .rename(columns={'side': 'n_unique_side'})
)
single_keys = side_counts[side_counts['n_unique_side'] == 1][keys]

side_map = (
    tmp.merge(single_keys, on=keys, how='inner')
    .drop_duplicates(subset=keys)[keys + ['side']]
)

# Merge side into merged df
df_full_with_side = df_merged.merge(
    side_map,
    left_on=['name', 'gameweek', 'season'],
    right_on=['team_name', 'gameweek', 'season'],
    how='left'
)

# Cleanup duplicate team_name columns if created
if 'team_name_x' in df_full_with_side.columns and 'team_name_y' in df_full_with_side.columns:
    df_full_with_side['team_name'] = df_full_with_side['team_name_x']
    df_full_with_side = df_full_with_side.drop(columns=['team_name_x', 'team_name_y'])

print('Merged shape:', df_merged.shape)
print('With side shape:', df_full_with_side.shape)
print('Ambiguous team-week-season keys:', int((side_counts['n_unique_side'] > 1).sum()))
print('Rows with missing side:', int(df_full_with_side['side'].isna().sum()))

df_full_with_side[['name', 'gameweek', 'season', 'side']].head()
```

```python
# === DROP RECOMMENDED TEAM-HEAVY COLUMNS (df_full_with_side) ===
drop_cols_team_excess = [
    # duplicate team identifiers (keep player-side 'name' key)
    'team',
    'team_name',

    # redundant xG/possession decomposition
    'xg_open_play',
    'xg_set_play',
    'own_half',
    'opposition_half',

    # noisy shot decomposition
    'shots_off_target',
    'blocked_shots',
    'hit_woodwork',

    # style descriptors
    'accurate_long_balls',
    'accurate_long_balls_pct',
    'accurate_crosses',
    'accurate_crosses_pct',

    # low-signal event noise
    'fouls_committed',
    'offsides',
    'throws',

    # overlapping duel stats
    'duels_won',
    'ground_duels_won',
    'ground_duels_won_pct',
    'aerial_duels_won',
    'aerial_duels_won_pct',

    # physical tracking block
    'distance_covered',
    'walking_distance',
    'running_distance',
    'sprinting_distance',
    'number_of_sprints',
    'top_speed',
]

# Optional strict pre-match setup:
# drop_cols_team_excess.append('score')

to_drop = [c for c in drop_cols_team_excess if c in df_full_with_side.columns]
df_model = df_full_with_side.drop(columns=to_drop).copy()

print('Dropped:', len(to_drop))
print('Before:', df_full_with_side.shape)
print('After :', df_model.shape)
print('Dropped columns:', to_drop)
```
